```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:/Users/USER/Downloads/archive/amazon.csv")
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹','')
```

```python
df['discounted_price'] = df['discounted_price'].str.replace(',','').astype('float')
```

```python
df.info()
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹','')
```

```python
df['actual_price'] = df['actual_price'].str.replace(',','').astype('float')
```

```python
df['discount_percentage']=df['discount_percentage'].str.replace('%','').astype('int')
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df['rating_count']=df['rating_count'].fillna(df['rating'].median())
```

```python
df['rating_count']=df['rating_count'].str.replace(',','')
```

```python
df['rating_count'].isnull().sum()
```

```python
df['rating_count']=df['rating_count'].astype('int')
```

```python
sns.histplot(df['discounted_price'], bins=30, kde=True)
plt.title("Distribution of Price")
plt.xlabel("Price")
plt.ylabel("Frequency")
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df['discount_percentage'], bins=30, kde=True)
plt.title("Distribution of discount percentage")
plt.xlabel("Discount %")
plt.ylabel("Frequency")
```

```python
df['category'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 categories")
plt.xlabel("Categories")
plt.ylabel("Frequency")
#plt.xticks(rotation=45)
```

```python
top_reviewed = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False).head(10)
top_reviewed.plot(kind='bar')
```
