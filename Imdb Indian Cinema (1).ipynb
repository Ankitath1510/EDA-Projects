```python
actors = df[['Actor 1', 'Actor 2', 'Actor 3']].melt(value_name='Actor')['Actor']

# Drop missing values
actors = actors.dropna()

# Count the frequency of actors
actor_collab = actors.value_counts().head(10)

# Create a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(pd.DataFrame(actor_collab), annot=True, cmap='Blues')
plt.title('Actor Collaboration Frequency (Top 10)')
plt.show()
```

```python
# Explode actor columns into a single column
actors = filtered_data[['Actor 1', 'Actor 2', 'Actor 3']].melt(value_name='Actor')['Actor']

# Drop missing values
actors = actors.dropna()

# Count the frequency of actors
actor_collab = actors.value_counts().head(10)

# Create a heatmap
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(pd.DataFrame(actor_collab), annot=True, cmap='Blues')
plt.title('Actor Collaboration Frequency (Top 10)')
plt.show()
```

```python
filtered_dir = df[df['Director'] != 'Unknown']
# Count movies by director
top_directors = filtered_dir['Director'].value_counts().head(10)

# Bar Plot for Top Directors
sns.barplot(x=top_directors.values, y=top_directors.index)
plt.title('Top 10 Directors with Most Movies')
plt.xlabel('Number of Movies')
plt.ylabel('Director')
plt.show()
```

```python
# Group by Year and calculate average rating
avg_rating_by_year = df.groupby('Year')['Rating'].mean().reset_index()

# Line Plot for Average Rating by Year
sns.lineplot(data=avg_rating_by_year, x='Year', y='Rating')
plt.title('Average Rating by Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.show()
```

```python
# Splitting genres and counting frequencies
genre_counts = df['Genre'].str.split(', ').explode().value_counts().head(10)

# Bar Plot for Top Genres
sns.barplot(x=genre_counts.index, y=genre_counts.values)
plt.title('Top Genres')
plt.xlabel('Genre')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

```python
# Sort the dataset by the 'Rating' column in descending order
top_movies = df.sort_values(by='Rating', ascending=False)

# Display the top 10 movies
print(top_movies[['Name', 'Year', 'Rating']].head(10))


# Get the top 10 movies
top_10_movies = top_movies[['Name', 'Rating']].head(10)

# Create the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(y=top_10_movies['Name'], x=top_10_movies['Rating'], palette='viridis')
plt.xlabel('Rating')
plt.ylabel('Movie Name')
plt.title('Top 10 Movies by Rating')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest-rated movie at the top
plt.show()
```
